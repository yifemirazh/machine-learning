```python
import networkx as nx
import matplotlib.pyplot as plt

# Initialize a directed graph for clarity (optional)
G = nx.Graph()

# Add nodes with roles
nodes = {
    'Firewall': {'role': 'security'},
    'Router': {'role': 'routing'},
    'Switch1': {'role': 'switching'},
    'Switch2': {'role': 'switching'},
    'Server1': {'role': 'application'},
    'Server2': {'role': 'application'},
    'DB_Server': {'role': 'database'},
    'Backup_Server': {'role': 'backup'},
}

# Add the nodes with attributes
for node, attr in nodes.items():
    G.add_node(node, **attr)

# Add weighted edges (costs or latencies)
edges = [
    ('Firewall', 'Router', 1),
    ('Router', 'Switch1', 2),
    ('Router', 'Switch2', 3),
    ('Switch1', 'Server1', 1),
    ('Switch2', 'Server2', 1),
    ('Server1', 'DB_Server', 2),
    ('Server2', 'DB_Server', 2),
    ('DB_Server', 'Backup_Server', 3),
    # Redundant links for fault tolerance
    ('Switch1', 'Switch2', 2),
    ('Router', 'DB_Server', 5),
]

# Add edges with weight attribute
for u, v, w in edges:
    G.add_edge(u, v, weight=w)

# Define node colors by role
role_colors = {
    'security': 'red',
    'routing': 'orange',
    'switching': 'skyblue',
    'application': 'green',
    'database': 'purple',
    'backup': 'gray'
}

# Get colors based on role
node_colors = [role_colors[G.nodes[n]['role']] for n in G.nodes]

# Position the nodes using spring layout
pos = nx.spring_layout(G, seed=42)

# Draw the graph
plt.figure(figsize=(12, 8))
nx.draw(G, pos, with_labels=True, node_color=node_colors, node_size=1800, font_size=10, font_weight='bold')
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)

plt.title("Advanced Network Graph with Redundancy and Security Roles")
plt.axis('off')
plt.show()
```
